```python
import pandas as pd 
```

```python
df = pd.read_csv("Sports Shoes.csv")
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
df["ratings"] = pd.to_numeric(df["ratings"], errors= "coerce")
df["ratings"] = df["ratings"].fillna(df["ratings"].mean())
print(df["ratings"])
```

```python
df["no_of_ratings"] = df["no_of_ratings"].str.replace(",", "", regex= False)
df["no_of_ratings"] = pd.to_numeric(df["no_of_ratings"], errors= "coerce")
df["no_of_ratings"] = df["no_of_ratings"].fillna(df["no_of_ratings"].mean())
print(df["no_of_ratings"])
```

```python
df["discount_price"] = df["discount_price"].str.replace("₹", "", regex=False)
df["discount_price"] = pd.to_numeric(df["discount_price"], errors="coerce")
df["discount_price"] = df["discount_price"].fillna(df["discount_price"].mean())
print(df["discount_price"])
```

```python
df["actual_price"] = df["actual_price"].str.replace("\u20B9", "", regex=False)
df["actual_price"] = df["actual_price"].str.replace(",", "", regex=False)
df["actual_price"] = pd.to_numeric(df["actual_price"], errors="coerce")
```

```python
print(df[df["actual_price"].isna()])
```

```python
print(df["actual_price"].head(20).tolist())
```

```python
df["actual_price"] = df["actual_price"].fillna(df["actual_price"].mean())
print(df["actual_price"])
```

```python
high_rated = df[df["ratings"] > 4.5]
print(high_rated)
```

```python
low_price = df[df["actual_price"] < 999] 
print(low_price)
```

```python
my_choice = df[(df["actual_price"] < 3999) & (df["ratings"] > 4.5) & (df["no_of_ratings"] > 50)]
print(my_choice)
```

```python
df.to_csv("sports_shoes_cleaned.csv", index=False)
```
